```python
from PIL import Image,ImageFilter
```

```python
img = Image.open('./pikachu.jpg')
```

```python
print(img)#PIL object for the image pikachu
```

```python
print(img.size,
img.format,
img.mode)
```

```python
dir(img)
```

```python
filtered_img = img.filter(ImageFilter.BLUR)
```

```python
filtered_img.save("pikachu_blur.png","png")# to save file and to change file type
```

```python
filtered_img = img.convert('L')
filtered_img.rotate(150)
```

```python
filtered_img.resize((300,300))#it takes tuple
```

```python
#To CROP
box = (0,0,600,400)
region = img.crop(box)
region
```

```python
#To change the size of image without squishing
filtered_img.thumbnail((400,400))#it keeps aspect ratio like size:400,400 same
```
